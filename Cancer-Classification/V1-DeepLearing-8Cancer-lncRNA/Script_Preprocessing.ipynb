```python
#zero elimination and knn imputation
import math
import pandas as pd
import numpy as np
from fancyimpute import KNN

df = pd.read_csv('Data/TCGA_LUAD_htseq_fpkm_lncRNA1.csv')
print("Raw data dimension:", df.shape)

data = df[((df.loc[:, df.columns != 'Ensembl_ID']).T != 0.0).any()]    #.all for atleast one zero
print("Dimension after row with all zero:", data.shape)

data = data.replace(0.0, np.nan)
col_threshold = int(math.ceil(0.1*data.shape[1]))
data = data.dropna(axis=0, how='all', thresh=col_threshold, subset=None, inplace=False)
print("Dimension after row with 10% zero:", data.shape)

data_knn = KNN(k=10).complete(data.iloc[0:,1:].values)
data_im = pd.DataFrame(data_knn)
Ensembl_IDs = pd.DataFrame(data.iloc[0:,0].values)
data = pd.concat([Ensembl_IDs, data_im], axis=1)

data.to_csv('Data/TCGA_LUAD_htseq_fpkm_lncRNA_small_eliminated.csv', sep=',', encoding='utf-8',index=False, header=False)
print("File has been written!")
```
